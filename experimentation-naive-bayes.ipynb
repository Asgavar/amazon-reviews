# Naive Bayes approach - manual experimentation

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import common

import sklearn.naive_bayes
```

## 1-X Occurence-counting vector representation

### 1-1 "Dirty" data

#### All classes at once prediction on randomly downsampled training set

```python
train_as_vec, train_targets, test_as_vec, test_targets = common.load_dataset('random_downsampling', 'count', None)
random_downsample_all_at_once_classifier = sklearn.naive_bayes.MultinomialNB(fit_prior=False)
random_downsample_all_at_once_classifier.fit(train_as_vec, train_targets)
```

```python
common.display_classifier_performance(random_downsample_all_at_once_classifier, test_as_vec, test_targets)
```

#### All classes at once on full training set, with **no weights**

```python
train_as_vec, train_targets, test_as_vec, test_targets = common.load_dataset('full', 'count', None)
full_all_at_once_classifier = sklearn.naive_bayes.MultinomialNB(fit_prior=False)
full_all_at_once_classifier.fit(train_as_vec, train_targets)
```

```python
common.display_classifier_performance(full_all_at_once_classifier, test_as_vec, test_targets)
```

#### All classes at once on full training set, **with weights**

```python
train_as_vec, train_targets, test_as_vec, test_targets = common.load_dataset('full', 'count', None)
weighted_full_all_at_once_classifier = sklearn.naive_bayes.MultinomialNB(fit_prior=True)
weighted_full_all_at_once_classifier.fit(train_as_vec, train_targets)
```

```python
common.display_classifier_performance(weighted_full_all_at_once_classifier, test_as_vec, test_targets)
```
